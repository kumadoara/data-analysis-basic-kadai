```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 折れ線グラフ
a = [5, 9, 1, 3, 4]
b= [2, 3, 7, 9, 10]

# sin関数
c = np.linspace(0,10,30)
d = 2 + 2 * np.sin(2 * c)

# ヒストグラム
e = np.random.rand(30)

# 散布図
f = np.random.rand(20)
g = np.random.rand(20)

fig, axes = plt.subplots(2, 3, tight_layout=True)

axes[0,0].plot(a,b)     # 折れ線グラフ
axes[0,1].plot(c,d)     # sin関数
axes[0,2].hist(e)       # ヒストグラム
axes[1,0].scatter(f,g)      # 散布図
plt.show()
```
